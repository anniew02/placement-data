#### Logisitic Regression Model
To address our question of interest, we used a logistic regression model. Logistic regression models are used to predict the probability of a binary outcome based on one or more predictor variables which is appropriate for our question as the response variable `PlacementStatus` is binary and there are multiple predictor variables which are numerical or categorical.

#### Model Assumptions
- Observations are independent of each other
- Linearity between logit of the response and the covariates
- No multicollinearity between the covariates

#### Limitations

### Model Selection
#### Lasso Model
We chose the LASSO model for logistic regression because it simultaneously performs feature selection and shrinks coefficients, effectively eliminating irrelevant variables. While Lasso is excellent for training and selection, it introduces bias in coefficient estimates due to regularization. To address this, we will apply post-LASSO debiasing by fitting an unpenalized logistic regression model using only the variables selected by LASSO. This approach allows us to obtain unbiased coefficient estimates, making the model suitable for accurate inference.

```r
# split the data
set.seed(875)

testing_data <- data %>%
  sample_frac(0.6, replace=FALSE) # experiment with this
training_data <- data %>%
  setdiff(testing_data)

head(training_data)
```

```r
training_data_lasso <- training_data |>
    mutate(PlacementStatus = if_else(PlacementStatus == "Placed", 1, 0)) |>
    mutate(PlacementTraining = if_else(PlacementTraining == "Yes",1,0)) |>
    mutate(ExtracurricularActivities = if_else(ExtracurricularActivities == "Yes", 1, 0))

testing_data_lasso <- testing_data |>
    mutate(PlacementStatus = if_else(PlacementStatus == "Placed", 1, 0)) |>
    mutate(PlacementTraining = if_else(PlacementTraining == "Yes",1,0)) |>
    mutate(ExtracurricularActivities = if_else(ExtracurricularActivities == "Yes", 1, 0))
```

```r
x <- dplyr::select(training_data_lasso, -PlacementStatus)  |>
     as.matrix()
y <- training_data_lasso$PlacementStatus
lasso_cv <- cv.glmnet(x = x, y = y, family = "binomial", alpha = 1)
```

```r
coef(lasso_cv, s = "lambda.min")
```

### Model Implementation
Now, we apply post-LASSO debiasing by fitting a least-squares logistic regression model with no penalization, using the variables selected by LASSO.

```r
selected_vars <- c("CGPA", "Projects", "WorkshopsCertifications", "AptitudeTestScore", 
                   "SoftSkillsRating", "ExtracurricularActivities", "PlacementTraining", 
                   "SSC_Marks", "HSC_Marks")

logistic_model <- glm(PlacementStatus ~ ., data = training_data[, c("PlacementStatus", selected_vars)], 
                      family = "binomial")

summary(logistic_model)
```

Now we will analyze the 95% confidence intervals of covariates included in our logistic model.

```r
coefs <- summary(logistic_model)$coefficients[, 1]
std_errs <- summary(logistic_model)$coefficients[, 2]

ci_lower <- coefs - 1.96 * std_errs
ci_upper <- coefs + 1.96 * std_errs

coef_df <- data.frame(
  Variable = names(coefs),
  Estimate = round(coefs,3),
  CI_Lower = round(ci_lower, 3),
  CI_Upper = round(ci_upper, 3)
)

coef_df
```

The results of the logistic regression model indicate that several variables are statistically significant predictors of placement status. CGPA, projects, aptitude test scores, soft skills rating, extracurricular activities, placement training, and both SSC and HSC marks have positive coefficients and 95% confidence intervals that do not include zero, suggesting that higher values in these areas increase the likelihood of placement. Specifically, placement training and extracurricular activities have relatively strong effects on placement outcomes. However, workshops and certifications did not show a significant effect, as its confidence interval included zero. These findings highlight the importance of academic performance, practical experience, soft skills, and preparation for placements in determining the likelihood of securing a placement.
